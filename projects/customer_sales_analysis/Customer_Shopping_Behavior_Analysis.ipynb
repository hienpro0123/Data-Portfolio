# üßπ Data Cleaning

```python
# Ki·ªÉm tra xem d·ªØ li·ªáu c√≥ b·ªã r·ªóng (Nan)
df.isnull().sum()
```

```python
# X√≥a d√≤ng b·ªã tr√πng l·∫∑p (v√¨ ch·ªâ c√≥ 1 d√≤ng)
df = df.dropna()

#Ki·ªÉm tra l·∫°i xem d·ªØ li·ªáu c√≤n Nan hay kh√¥ng
df.isnull().sum()
```

```python
# Ki·ªÉm tra duplicate to√†n b·ªô d√≤ng
duplicate_rows = df[df.duplicated()]
print(f"S·ªë d√≤ng duplicate to√†n b·ªô: {duplicate_rows.shape[0]}")
```

```python
#Chu·∫©n h√≥a ki·ªÉu d·ªØ li·ªáu c·ªôt "Invoice Date" sang datetime
df['Invoice Date'] = pd.to_datetime(df['Invoice Date'], format='%m/%d/%Y')
df.info()
```

```python
# Chu·∫©n h√≥a c·ªôt "Age" v√† "Quantity" sang ki·ªÉu d·ªØ li·ªáu int
df['Age'] = df['Age'].astype("Int32")
df['Quantity'] = df['Quantity'].astype("Int32")
```

```python
# T·∫°o nh√£n nh√≥m tu·ªïi
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']

#T·∫°o c·ªôt 'Age Group'
df['Age Group'] = pd.qcut(df['Age'],q=4,labels=labels)

# Hi·ªÉn th·ªã
df.head()
```

```python
# Th√™m 1 c·ªôt "Total Price"
df['Total Price'] = df['Quantity'] * df['Price']

df["Total Price"] = df["Total Price"].round(2)

# Hi·ªÉn th·ªã
df.head()
```

```python
# T√¨m gi√° tr·ªã outlier c·ªßa c·ªôt "Total Price"
col = 'Total Price'
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# C√°c d√≤ng b·ªã coi l√† outlier
outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
print("S·ªë l∆∞·ª£ng outlier:", outliers.shape[0])

# Xem v√†i outlier ƒë·∫ßu ti√™n
outliers[[col]].head()
```

```python
import pyodbc

cursor = conn.cursor()

# 1) Ki·ªÉm tra b·∫£ng ƒë√£ t·ªìn t·∫°i ch∆∞a
check_sql = """
SELECT COUNT(*) 
FROM sys.tables 
WHERE name = 'data_cleaned' AND schema_id = SCHEMA_ID('dbo');
"""

cursor.execute(check_sql)
exists = cursor.fetchone()[0]  # 0 = ch∆∞a c√≥, >0 = ƒë√£ t·ªìn t·∫°i

if exists > 0:
    print("‚ÑπÔ∏è B·∫£ng 'data_cleaned' ƒë√£ t·ªìn t·∫°i, b·ªè qua b∆∞·ªõc t·∫°o b·∫£ng.")
else:
    # 2) N·∫øu ch∆∞a c√≥ th√¨ t·∫°o b·∫£ng
    sql_create = """
    CREATE TABLE dbo.data_cleaned (
        Invoice_Number NVARCHAR(20),
        Customer_ID    NVARCHAR(20),
        Gender         NVARCHAR(10),
        Age            INT,
        Category       NVARCHAR(50),
        Quantity       INT,
        Price          DECIMAL(18,2),
        Payment_Method NVARCHAR(50),
        Invoice_Date   DATE,
        Shopping_Mall  NVARCHAR(50),
        Age_Group      NVARCHAR(20),
        Total_Price    DECIMAL(18,2)
    );
    """

    cursor.execute(sql_create)
    conn.commit()
    print("‚úÖ ƒê√£ t·∫°o m·ªõi b·∫£ng 'data_cleaned'.")
```

```python
# L·∫•y danh s√°ch Invoice_Number ƒë√£ c√≥ trong b·∫£ng
cursor.execute("SELECT DISTINCT Invoice_Number FROM data_cleaned")
existing_invoices = {row[0] for row in cursor.fetchall()} 

print(f"üîé ƒê√£ c√≥ s·∫µn {len(existing_invoices)} Invoice_Number trong b·∫£ng data_cleaned.")

# Insert d·ªØ li·ªáu v√†o
insert_sql = """
INSERT INTO data_cleaned (
    Invoice_Number,
    Customer_ID,
    Gender,
    Age,
    Category,
    Quantity,
    Price,
    Payment_Method,
    Invoice_Date,
    Shopping_Mall,
    Age_Group,
    Total_Price
)
VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
"""

# T·∫°o list tuple t·ª´ df, nh∆∞ng ch·ªâ l·∫•y nh·ªØng Invoice_Number CH∆ØA c√≥ trong b·∫£ng
rows_to_insert = []
for _, row in df.iterrows():
    invoice_num = row["Invoice Number"]   # l·∫•y tr∆∞·ªõc cho d·ªÖ d√πng

    # N·∫øu invoice ƒë√£ t·ªìn t·∫°i trong b·∫£ng th√¨ b·ªè qua
    if invoice_num in existing_invoices:
        # print(f"B·ªè qua Invoice_Number ƒë√£ t·ªìn t·∫°i: {invoice_num}")
        continue

    rows_to_insert.append(
        (
            invoice_num,
            row["Customer ID"],
            row["Gender"],
            int(row["Age"]),
            row["Category"],
            int(row["Quantity"]),
            float(row["Price"]),
            row["Payment Method"],
            pd.to_datetime(row["Invoice Date"]).date(),
            row["Shopping Mall"],
            row["Age Group"],
            float(row["Total Price"])
        )
    )

# N·∫øu c√≤n d√≤ng m·ªõi th√¨ insert
if rows_to_insert:
    cursor.executemany(insert_sql, rows_to_insert)
    conn.commit()
    print(f"‚úÖ ƒê√£ insert {len(rows_to_insert)} d√≤ng M·ªöI v√†o b·∫£ng data_cleaned.")
else:
    print("‚ö†Ô∏è Kh√¥ng c√≥ d√≤ng m·ªõi n√†o ƒë·ªÉ insert (t·∫•t c·∫£ Invoice_Number ƒë·ªÅu ƒë√£ t·ªìn t·∫°i).")
```
